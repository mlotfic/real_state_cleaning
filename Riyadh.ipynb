## Cleaning process

```python
# rename first column to month
df_Riyadh = df_Riyadh.rename(columns ={df_Riyadh.columns[0] : 'month'})
df_Riyadh 
```

```python
# create empty list to save the indexs of district name 
list_redundant =[]

# create a range of number for do the operation at every row in dataset
x = range(8, df_Riyadh.shape[0], 1)


for i in x:
    # check if it is district then save this didtric name and it's index
    if (df_Riyadh.iloc [i, 0] not in months_names):
        print(df_Riyadh.iloc [i, 0], '-----', df_Riyadh.index[i])
        temp = df_Riyadh.iloc [i, 0]
        list_redundant.append(df_Riyadh.index[i])  
        
    # else if it is a month set its district name                      
    elif (df_Riyadh.iloc [i, 0] in months_names):
            df_Riyadh.iloc [i, 1] = temp
```

## Defining some repeated code 

```python
def year_df(df, yr = "2018", lst = [0,1,2,3,4]):
    """
    input: orignal dataset , set the year of data extracted , list of columns the information to be extracted from
    rturn: dataframe with column names year, month, realstate_cls, realstate_type, price_riyals, area_m2, no_deals
    """
    # select the columns from orignal dataset
    df = df.iloc[:, lst]
    
    # Changing columns name with index number
    df = df.rename(
        columns ={
            df.columns[2]: 'price_riyals',
            df.columns[3]: 'area_m2',
            df.columns[4]: 'no_deals'
        })
    
    # set the year for data extracted
    df.insert(loc = 0, column= 'year', value= yr)
    
    # due to removing some row
    df.reset_index(inplace = True)
    
    # returning the processed data 
    return df.drop(df.columns[0], axis=1)
```

### 1- Get 2018 data from orignal database

```python
df = year_df(df_Riyadh, yr = "2018", lst = [0,1,2,3,4])
df.head(20)
```

### 2-Get 2019 data from orignal database

```python
df_2019 = year_df(df_Riyadh, yr = "2019", lst = [0,1,5,6,7])
df_2019.head(20)
```

### 3-Get 2020 data from orignal database

### 4-Get 2021 data from orignal database

```python
df_2021 = year_df(df_Riyadh, yr = "2021", lst = [0,1,11,12,13])
df_2021.head(20)
```

### Convert column to appropriate data type 

```python
# السعر بالريال السعودي
# fill none value with zeros
df['price_riyals'] = df['price_riyals'].fillna(0).astype(np.int64)

# المساحة بالمتر المربع
# fill none value with zeros
df['area_m2'] = df['area_m2'].fillna(0.0).astype(float)

# عدد الصفقات
# fill none value with zeros
df['no_deals'] = df['no_deals'].fillna(0).astype(int)
```

### Adding addtional information to database

```python
# adding city name
df.insert(loc = 2, column= 'city', value= df_city_name)

# adding تصنيف العقار
df.insert(loc = 3, column= 'realstate_cls', value= '')

# adding نوع العقار
df.insert(loc = 4, column= 'realstate_type', value= '')

# adding حالة العملية(معتمدة/ملغية/..)
df.insert(loc = 7, column= 'status', value = df_status)

# fill nan in string with empty 
df.fillna('', inplace=True)
```
